```python
#converting durestion_ms to seconds.
df_tracks["duration"]=df_tracks["duration_ms"].apply(lambda x:round(x/1000))
df_tracks.drop("duration_ms", inplace=True, axis=1)
```

# Visualizations

```python
corr_df=df_tracks.drop(["key","mode","explicit"],axis=1).corr(method="pearson")
plt.figure(figsize=(14,6))
heatmap=sns.heatmap(corr_df,annot=True,fmt=".1g", vmin=-1, vmax=1, center=0, cmap="inferno", linewidths=1,linecolor="Black")
heatmap.set_title("Correlation Heatmap between variable")
heatmap.set_xticklabels(heatmap.get_xticklabels(),rotation=90)
```

By this visualization, we get to know that acousticness have a strong negative correlation with energy. There is moderately strong positive relation between loudness and popularity, there is also moderately strong positive relation between danceability and valence.

```python
sample_df = df_tracks.sample(int(0.004*len(df_tracks)))
```

```python
print(len(sample_df))
```

Here, .4% of total dataset is 2346.

```python
#regression plot between loudness and energy
plt.figure(figsize=(10,6))
sns.regplot(data=sample_df, y = "loudness", x="energy", color="c").set(title="Loudness vs Energy Correlation")
```

Here, we can see that there is a very high correlation between loudness and energy,as the regression line is increasing gradually. Hence, energy increases, loudness of the song increases, energy decreases, loudness of song also decreases.

```python
#regression plot between loudness and energy
plt.figure(figsize=(10,6))
sns.regplot(data=sample_df, y = "popularity", x="acousticness", color="b").set(title="Popularity vs Acousticness Correlation")
```

Here, we see that when acousticness of song increases, the popularity of song decreases and if acousticness of song decreases, then popularity of song increases. Hence, we see that regression line gradually decreases.

```python
df_tracks['dates']=df_tracks.index.get_level_values('release_date')
df_tracks.dates=pd.to_datetime(df_tracks.dates) 
years=df_tracks.dates.dt.year
```

# Distribution Plot

Visualizing total number of songs in each year since 1922 that is available on Spotify

```python
#pip install --user seaborn==0.11.0
```

```python
sns.displot(years,discrete=True, aspect=2,height=5, kind="hist" ).set(title="Number of songs per year")
```

Here, number of songs for each year in the dataset according to the release dates have increased in recent years.

```python
#duration of songs over the years
total_dr = df_tracks.duration
fig_dims = (18,7)
fig, ax = plt.subplots(figsize=fig_dims)
fig=sns.barplot(x=years, y=total_dr, ax = ax, errwidth=False).set(title="Year vs Duration")
plt.xticks(rotation=90)
```

Here, we can see that duration of the songs in the early 20's was less, later the duration increased and remained constant till the year 2010, after 2010 the duration of the gradually started decreasing.

```python
df_genre =pd.read_csv("SpotifyFeatures.csv")
```

```python
df_genre.head(5)
```

```python
#duration of songs for different genres
plt.title("Duration of songs in different genres")
sns.color_palette("rocket", as_cmap=True)
sns.barplot(y="genre", x="duration_ms", data=df_genre)
plt.xlabel("Duration in milli seconds")
plt.ylabel("Genres")
```

Here, we see that the graph for classical genre and world genre, durations of the songs are more compared to other genres.

```python
#TOP 5 genres by popularity
sns.set_style(style="darkgrid")
plt.figure(figsize=(10,5))
top_genre=df_genre.sort_values("popularity", ascending=False).head(10)
sns.barplot(y="genre", x="popularity", data=top_genre).set(title="Top 5 Genres by popularity")
```
